```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
## sentences
sent = [
    "the glass of milk",
    "the glass of juice",
    "the cup of tea",
    "I am a good boy",
    "I am a good developer",
    "understand the meaning of words",
    "your videos are good"
]
```

```python
## define the vocabulary size
voc_size = 10000
```

```python
## One hot representation for every words
one_hot_repr = [one_hot(words, voc_size) for words in sent]
one_hot_repr
```

```python
## Word embedding representation
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np 
```

```python
## set max sentence length cuz the model needs to have same number of imputs for each timestamp
sent_len = 8
embedded_docs = pad_sequences(one_hot_repr, padding= "pre", maxlen= sent_len)
embedded_docs
```

```python
## feature reprensentation
dim = 10
```

```python
## Create embedding model
model = Sequential()
model.add(Embedding(voc_size, dim, input_length = sent_len))
model.compile('adam' , 'mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
