# Support Vector Machine Implementation

## Support Vector Classifier SVC

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Creating synthetic data points
from sklearn.datasets import make_classification
```

```python
X,y = make_classification(n_samples=1000, n_features=2, n_classes=2, n_clusters_per_class=2, n_redundant=0, random_state= 14)
```

```python
X
```

```python
y
```

```python
X = pd.DataFrame(X)
```

```python
sns.scatterplot(x=X[0],y=X[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=12)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC(kernel='linear')
svc.fit(X_train, y_train)
```

```python
y_pred = svc.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, y_pred) , confusion_matrix(y_test, y_pred))
```

```python
rbf = SVC()
rbf.fit(X_train, y_train)
y_pred = rbf.predict(X_test)
```

```python
print(classification_report(y_test, y_pred) , confusion_matrix(y_test, y_pred))
```

### Hyperparameter Tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {'C': [0.1,1,10,100,100],
              'gamma': [1,0.1,0.01,0.001,0.0001],
              'kernel': ['rbf' , 'linear' , 'sigmoid' , 'poly']}
```

```python
grid = GridSearchCV(SVC() , param_grid= param_grid, refit = True, cv = 5, verbose=3)
```

```python
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred_grid = grid.predict(X_test)
```

```python
print(classification_report(y_test, y_pred_grid) , confusion_matrix(y_test, y_pred_grid))
```

```python
## Feature Encoding: Label Encodig, One Hot Encoding
X = df[['tip', 'sex', 'smoker', 'day', 'time', 'size']]
y = df['total_bill']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25, random_state=98)
```

```python
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
```

```python
le1,le2,le3 = LabelEncoder(), LabelEncoder(), LabelEncoder()
```

```python
import warnings
warnings.filterwarnings('ignore')
X_train['sex'] = le1.fit_transform(X_train['sex'])
X_train['smoker'] = le2.fit_transform(X_train['smoker'])
X_train['time'] = le3.fit_transform(X_train['time'])
```

```python
X_train.head()
```

```python
X_test['sex'] = le1.transform(X_test['sex'])
X_test['smoker'] = le2.transform(X_test['smoker'])
X_test['time'] = le3.transform(X_test['time'])
X_test.head()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ct = ColumnTransformer(transformers=[('onehot' , OneHotEncoder(drop='first'), [3])], remainder= 'passthrough')
```

```python
import sys
np.set_printoptions(threshold=sys.maxsize)
X_train = ct.fit_transform(X_train)
X_test = ct.transform(X_test)
```

```python
# SVR
from sklearn.svm import SVR
```

```python
svr = SVR()
svr.fit(X_train, y_train)
y_pred = svr.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
print(r2, mae)
```
