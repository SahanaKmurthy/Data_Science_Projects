# Decision Tree Classifier

```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
print(iris['DESCR'])
```

```python
X = pd.DataFrame(iris['data'], columns= ['sepal length' , 'sepal width' , 'petal length' , 'petal width'])
X.head()
```

```python
y = iris['target']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.2, random_state= 12)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
treeclassifier = DecisionTreeClassifier()
treeclassifier.fit(X_train, y_train)
```

```python
# Visualize decision tree
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(treeclassifier, filled = True)
```

## Post Pruning

```python
treeclassifier2 = DecisionTreeClassifier(max_depth= 2)
treeclassifier2.fit(X_train, y_train)
```

```python
y_pred = treeclassifier2.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)
print(cm , cr)
```

## Pre Pruning and Hyperparamter Tuning

```python
params = {'criterion': ['gini' , 'entropy' , 'log_loss'] ,
          'splitter' : ['best' , 'random'],
          'max_depth' : [1,2,3,5,10],
          'max_features' : ['auto' , 'sqrt' , 'log2']
          }
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(treeclassifier , param_grid= params, cv = 5 , scoring= 'accuracy')
```

```python
import warnings
warnings.filterwarnings("ignore")
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
cm = confusion_matrix(y_test, y_pred)
cr = classification_report(y_test, y_pred)
print(cm , cr)
```

# Decision Tree Regressor

```python
from sklearn.datasets import load_diabetes
```

```python
dataset = load_diabetes()
dataset
```

```python
print(dataset['DESCR'])
```

```python
X = pd.DataFrame(dataset['data'] , columns = ['age' , 'sex' , 'bmi' , 'bp' , 's1' , 's2' , 's3' , 's4' , 's5' , 's6'])
y = dataset['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size= 0.3, random_state=88)
```

```python
## correlation
X_train.corr()
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(), annot= True)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
reg = DecisionTreeRegressor()
reg.fit(X_train, y_train)
```

## Hyperparamter Tuning

```python
params = {
    'criterion' : ['squared_error' , 'friedman_mse' , 'absolute_error'],
    'splitter' : ['best' , 'random'],
    'max_depth' : [1,2,3,5,10,20,50,100],
    'max_features' : ['auto' , 'sqrt' , 'log2']
}
```

```python
grid = GridSearchCV(reg, param_grid= params, cv = 5, scoring= 'neg_mean_squared_error')
```

```python
warnings.filterwarnings("ignore")
grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
y_pred = grid.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
mse = mean_squared_error(y_test , y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(mse, mae, r2)
```

```python
best_model = grid.best_estimator_
```

```python
# Visualize Decision Tree
plt.figure(figsize=(15,10))
tree.plot_tree(best_model, filled = True)
```
