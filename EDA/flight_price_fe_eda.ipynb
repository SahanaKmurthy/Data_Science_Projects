```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_excel("../data/flight_price.xlsx")
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['Date'] = df['Date_of_Journey'].str.split("/").str[0].astype(int)
df['Month'] = df['Date_of_Journey'].str.split("/").str[1].astype(int)
df['Year'] = df['Date_of_Journey'].str.split("/").str[2].astype(int)
df.head()
```

```python
df.info()
```

```python
df.drop('Date_of_Journey' , axis = 1 , inplace=True)
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x: x.split(" ")[0])
```

```python
df['arrival_hour'] = df['Arrival_Time'].str.split(":").str[0].astype(int)
df['arrival_min'] = df['Arrival_Time'].str.split(":").str[1].astype(int)
df.head()
```

```python
df.drop('Arrival_Time' , axis = 1 , inplace = True)
```

```python
df['departure_hour'] = df['Dep_Time'].str.split(":").str[0].astype(int)
df['departure_min'] = df['Dep_Time'].str.split(":").str[1].astype(int)
df.head()
```

```python
df.drop("Dep_Time" , axis = 1 , inplace=True)
```

```python
df['Total_Stops'].unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df['Total_Stops'] = df['Total_Stops'].map({'non-stop':0 , '1 stop' : 1 , '2 stops': 2 , '3 stops' : 3 , '4 stops' : 4 , np.nan:1})
```

```python
df.head()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.drop('Route' , axis = 1 , inplace=True)
```

```python
df['duration_hour'] = df['Duration'].apply(lambda x: 0 if x.split(" ")[0].find('h') == -1 else x.split(" ")[0].split("h")[0]).astype(int)
df['duration_min'] = df['Duration'].str.split(" ").str[1].str.split("m").str[0]
```

```python
df.describe()
```

```python
df['duration_min'].unique()
```

```python
df[df['duration_min'].isna()]
```

```python
df['duration_min'] = df['duration_min'].fillna("0").astype(int)
```

```python
df.head()
```

```python
df.drop('Duration' , axis = 1 , inplace=True)
```

```python
df['Airline'].unique()
```

```python
df['Source'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder()
array = encoder.fit_transform(df[['Airline' , 'Source' , 'Destination']]).toarray()
```

```python
ohe_df = pd.DataFrame(array , columns = encoder.get_feature_names_out())
```

```python
ohe_df.head()
```

```python
merged_df = pd.concat([df , ohe_df] , axis = 1)
merged_df.head()
```

```python
merged_df.drop(['Airline' , 'Source' , 'Destination' , 'Additional_Info'] , axis = 1 , inplace=True)
merged_df.head()
```
